### Design and discussion filtering algorithm

#### 1. Remove magical functions
Magic functions, typically starting and ending with double underscores (e.g., __init__, __call__), typically handle low-level Python mechanics or object manipulations rather than core application logic.  By filtering these functions out, we simplify data sets, making it easier to understand high-level processes on model training, optimization steps, and data transformations. Magic functions are always repeated at a very high frequency. Removing them prevents the distraction caused by trivial calls and unnecessary processing.   

```python 
events = events.filter(~traces["function"].str.contains(r"\.__*__")) 
```

#### 2. Remove private or inner functions
FAET Relation mainly focus on high level ordered function calls group. Private and internal functions often starting with a single underscore, handle low-level operations or helper tasks (many helper function of API), which are not directly relevant to the core logic.

```python
events = events.filter(~traces["function"].str.contains(r"\._")) 
```

```python
    threshold = 6

        (events["function"] != events["function"].shift()).cumsum().alias("group_id")
    )

    group_counts = events.groupby("group_id").agg([
        pl.col("function").first().alias("function"),
        pl.count("function").alias("count")
    ])

    functions_to_remove = group_counts.filter(pl.col("count") > threshold)["function"]

    events = events.filter(~events["function"].is_in(functions_to_remove))
    function_pool = set(events["function"].unique().to_list())
```

Here we set threshold to be 6 (need to be futhur explored if appropriate). This means we allow at most 6 consecutive calls of a single function (including both "pre" and "post").
